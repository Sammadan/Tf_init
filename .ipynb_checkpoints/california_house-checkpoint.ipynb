```python
model=Sequential()
model.add(Dense(32, input_shape=(8,),activation='relu', kernel_initializer='normal'))
model.add(Dense(32, kernel_initializer='normal'))
model.add(Dense(1))
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=1000, batch_size=41)
```

```python
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(build_fn=model, epochs=50, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = KFold(n_splits=10, random_state=7)
results = cross_val_score(pipeline, X_train, y_train, cv=kfold)
print("Larger: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
